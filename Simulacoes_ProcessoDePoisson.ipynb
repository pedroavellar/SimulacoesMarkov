```
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

# Exercício 1

```
# Função para simular um Processo de Poisson
# rate é a taxa, tmin o tempo inicial, tmax o tempo final, n_sim o número de simulações
def sim_poisson(rate, tmin, tmax, n_sim):
  Ns = []
  for s in range(0, n_sim):
      te = tmin
      ne = 0
      while (te < tmax):
          te = te + np.random.exponential(1/rate)
          if(te < tmax):
              ne = ne + 1
      Ns.append(ne)
  # Plot do histograma de N x P(N)
  plt.figure(figsize=(10,6))
  bins = range(np.min(Ns), np.max(Ns))
  plt.hist(x=Ns, bins=bins, color='blue', alpha=0.7, rwidth=0.75, density=True, label='Simulacao')
  plt.xlabel('N(t)', fontsize=20)
  plt.ylabel('P[N(t)]', fontsize=20)
  # Cria a curva teórica
  x = range(np.min(Ns),np.max(Ns))
  P = []
  for k in x:
      prob = (np.exp(-rate*(tmax-tmin))*(rate*(tmax-tmin))**k)/math.factorial(k)
      P.append(prob)
  plt.plot(x,P, color = 'red', label='Curva Teorica')
  plt.legend()
  plt.xticks(bins)
  plt.show(True)
```

```
sim_poisson(rate = 5, tmin = 9, tmax = 11, n_sim = 100)
```

```
sim_poisson(rate = 5, tmin = 9, tmax = 11, n_sim = 1000)
```

# Exercício 4

```
rates = [2, 4] # Taxas (lambdas)
tmax = 10 # Tempo máximo

plt.figure(figsize=(10,6))
for rate in rates:
  N = []
  T = []
  ne = 0
  te = 0
  while(te < tmax):
    te = te + np.random.exponential(1/rate)
    if(te < tmax):
      ne = ne + 1
    N.append(ne)
    T.append(te)
  plt.step(T, N, label = "Lambda = %d"%rate)

plt.legend()
plt.xlabel('t', fontsize=20)
plt.ylabel('N(t)', fontsize=20)
plt.show(True)
```
